## 데이터 제외하기 dropna

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

```python
df.dropna(inplace=True) #전체 데이터 중에서 NaN을 포함하는 데이터 삭제
```

```python
df
```

- axis : index or columns  
- how : any or all

```python
df.dropna(axis='index', how='any') #NaN 가 하나라도 있는 row 삭제
```

```python
df.dropna(axis='columns') #NaN이 하나라도 있는 column 삭제
```
